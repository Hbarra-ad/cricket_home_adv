```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("../data/england_match_results.csv")
df.head()
```

```python
df['Date'] = df['Date'].astype(str).str[-4:]
df.head()
```

```python
#Replace one or more digits from the series column to just get the teams playing
#https://medium.com/@rgr5882/100-days-of-data-science-day-39-removing-unwanted-characters-from-text-columns-5b226a1f2fce#:~:text=The%20regex%20pattern%20%5B%5Ea%2D,removing%20them%20from%20the%20text.
df['Series'] = df['Series'].str.replace(r'[^a-zA-Z\s]', '', regex = True)
df.head()
```

```python
df['Home Team'] = df['Series'].str.split().str[0]
df.head()
```

```python
df['Result'] = df['Result'].str.split().str[:2].str.join(' ')
```

```python
df['Winner'] = df['Result'].str.split().str[:1].str.join(' ')
```

```python
df.head()
```

```python
df['Winner'] = df['Winner'].str.replace('Match', 'Draw', regex = False)
df.head()
```

```python
df = df.drop('Result', axis = 1)
df.head()
```

```python
df = df.drop('Test Number', axis = 1)
df.head()
```

```python
df = df.drop('Series', axis = 1)
```

```python
df.head()
```

```python
df = df[df['Home Team'] == 'England']
```

```python
df.head()
```

```python
df.to_csv('../data/clean_england_data.csv', index=False)
```
