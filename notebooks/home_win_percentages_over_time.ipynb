```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df_ind = pd.read_csv("../data/clean_india_data.csv")
df_eng = pd.read_csv("../data/clean_england_data.csv")
df_aus = pd.read_csv("../data/clean_australia_data.csv")
df_nz = pd.read_csv("../data/clean_new_zealand_data.csv")
df_sa = pd.read_csv("../data/clean_south_africa_data.csv")
```

```python
df_ind['Date'] = pd.to_datetime(df_ind['Date'], format='%Y')

df_ind['Decade'] = (df_ind['Date'].dt.year // 10) * 10

ind_no_draws = df_ind[df_ind['Winner'] != 'Draw']

ind_decade = ind_no_draws.groupby('Decade').size()
ind_wins_by_decade = ind_no_draws[ind_no_draws['Winner'] == 'India'].groupby('Decade').size()

ind_percentage = (ind_wins_by_decade / ind_decade) * 100
ind_percentage = ind_percentage.fillna(0)

print(ind_percentage)
```

```python
df_eng['Date'] = pd.to_datetime(df_eng['Date'], format='%Y')

df_eng['Decade'] = (df_eng['Date'].dt.year // 10) * 10

eng_no_draws = df_eng[df_eng['Winner'] != 'Draw']

eng_decade = eng_no_draws.groupby('Decade').size()
eng_wins_by_decade = eng_no_draws[eng_no_draws['Winner'] == 'England'].groupby('Decade').size()

eng_percentage = (eng_wins_by_decade / eng_decade) * 100
eng_percentage = eng_percentage.fillna(0)

print(eng_percentage)
```

```python
df_aus['Date'] = pd.to_datetime(df_aus['Date'], format='%Y')

df_aus['Decade'] = (df_aus['Date'].dt.year // 10) * 10

aus_no_draws = df_aus[df_aus['Winner'] != 'Draw']

aus_decade = aus_no_draws.groupby('Decade').size()
aus_wins_by_decade = aus_no_draws[aus_no_draws['Winner'] == 'Australia'].groupby('Decade').size()

aus_percentage = (aus_wins_by_decade / aus_decade) * 100
aus_percentage = aus_percentage.fillna(0)

print(aus_percentage)
```

```python
df_nz['Date'] = pd.to_datetime(df_nz['Date'], format='%Y')

df_nz['Decade'] = (df_nz['Date'].dt.year // 10) * 10

nz_no_draws = df_nz[df_nz['Winner'] != 'Draw']

nz_decade = nz_no_draws.groupby('Decade').size()
nz_wins_by_decade = nz_no_draws[nz_no_draws['Winner'] == 'New'].groupby('Decade').size()

nz_percentage = (nz_wins_by_decade / nz_decade) * 100
nz_percentage = nz_percentage.fillna(0)

print(nz_percentage)
```

```python
df_sa['Date'] = pd.to_datetime(df_sa['Date'], format='%Y')

df_sa['Decade'] = (df_sa['Date'].dt.year // 10) * 10

sa_no_draws = df_sa[df_sa['Winner'] != 'Draw']

sa_decade = sa_no_draws.groupby('Decade').size()
sa_wins_by_decade = sa_no_draws[sa_no_draws['Winner'] == 'South'].groupby('Decade').size()

sa_percentage = (sa_wins_by_decade / sa_decade) * 100
sa_percentage = sa_percentage.fillna(0)

print(sa_percentage)
```

```python
plt.figure(figsize=(12, 8))

plt.plot(ind_percentage.index, ind_percentage.values, label='India', color='orange')
plt.plot(eng_percentage.index, eng_percentage.values, label='England', color='blue')
plt.plot(aus_percentage.index, aus_percentage.values, label='Australia', color='gold')
plt.plot(nz_percentage.index, nz_percentage.values, label='New Zealand', color='black')
plt.plot(sa_percentage.index, sa_percentage.values, label='South Africa', color='green')

plt.title("Home Win Percentages Over Time by Country", fontsize=16)
plt.xlabel("Decade", fontsize=14)
plt.ylabel("Home Win Percentage", fontsize=14)
plt.legend(title="Teams")
plt.xticks(rotation=90)

plt.grid(True)

plt.show()
```
