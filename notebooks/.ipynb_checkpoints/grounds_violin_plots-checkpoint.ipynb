```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df_ind = pd.read_csv("../data/clean_india_data.csv")
df_eng = pd.read_csv("../data/clean_england_data.csv")
df_aus = pd.read_csv("../data/clean_australia_data.csv")
df_nz = pd.read_csv("../data/clean_new_zealand_data.csv")
df_sa = pd.read_csv("../data/clean_south_africa_data.csv")
```

```python
df_clean_ind = df_ind[df_ind['Winner'] != 'Draw']

df_clean_ind['Home Win'] = df_clean_ind['Winner'].apply(lambda x: 1 if x == 'India' else 0)

ind_matches = df_clean_ind.groupby('Ground').size()

ind_5_matches = ind_matches[ind_matches > 6].index
df_clean_ind = df_clean_ind[df_clean_ind['Ground'].isin(ind_5_matches)]

plt.figure(figsize=(12, 8))
sns.violinplot(x='Ground', y='Home Win', data=df_clean_ind)

plt.title('India Home Win Percentages by Ground (More than 6 Matches)', fontsize=16)
plt.xlabel('Ground', fontsize=14)
plt.ylabel('Home Win (1 = Win, 0 = Loss)', fontsize=14)
plt.xticks(rotation=90)
plt.show()
```
