```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df_ind = pd.read_csv("../data/clean_india_data.csv")
df_eng = pd.read_csv("../data/clean_england_data.csv")
df_aus = pd.read_csv("../data/clean_australia_data.csv")
df_nz = pd.read_csv("../data/clean_new_zealand_data.csv")
df_sa = pd.read_csv("../data/clean_south_africa_data.csv")
```

```python
ind_total_matches = len(df_ind)
ind_home_wins = len(df_ind[df_ind['Winner'] == 'India'])
ind_drawn_matches = len(df_ind[df_ind['Winner'] == 'Draw'])
ind_non_drawn_matches = ind_total_matches - ind_drawn_matches
ind_win_percentage = (ind_home_wins / ind_non_drawn_matches) * 100

print("Total Home Matches Played:", ind_total_matches)
print("Total Non Drawn Matches:", ind_non_drawn_matches)
print("Total Home Wins:", ind_home_wins)
print("Win Percentage:", ind_win_percentage, "%")
```

```python
eng_total_matches = len(df_eng)
eng_home_wins = len(df_eng[df_eng['Winner'] == 'England'])
eng_drawn_matches = len(df_eng[df_eng['Winner'] == 'Draw'])
eng_non_drawn_matches = eng_total_matches - eng_drawn_matches
eng_win_percentage = (eng_home_wins / eng_non_drawn_matches) * 100

print("Total Home Matches Played:", eng_total_matches)
print("Total Non Drawn Matches:", eng_non_drawn_matches)
print("Total Home Wins:", eng_home_wins)
print("Win Percentage:", eng_win_percentage, "%")
```

```python
aus_total_matches = len(df_aus)
aus_home_wins = len(df_aus[df_aus['Winner'] == 'Australia'])
aus_drawn_matches = len(df_aus[df_aus['Winner'] == 'Draw'])
aus_non_drawn_matches = aus_total_matches - aus_drawn_matches
aus_win_percentage = (aus_home_wins / aus_non_drawn_matches) * 100

print("Total Home Matches Played:", aus_total_matches)
print("Total Non Drawn Matches:", aus_non_drawn_matches)
print("Total Home Wins:", aus_home_wins)
print("Win Percentage:", aus_win_percentage, "%")
```

```python
nz_total_matches = len(df_nz)
nz_home_wins = len(df_nz[df_nz['Winner'] == 'New'])
nz_drawn_matches = len(df_nz[df_nz['Winner'] == 'Draw'])
nz_non_drawn_matches = nz_total_matches - nz_drawn_matches
nz_win_percentage = (nz_home_wins / nz_non_drawn_matches) * 100

print("Total Home Matches Played:", nz_total_matches)
print("Total Non Drawn Matches:", nz_non_drawn_matches)
print("Total Home Wins:", nz_home_wins)
print("Win Percentage:", nz_win_percentage, "%")
```

```python
sa_total_matches = len(df_sa)
sa_home_wins = len(df_sa[df_sa['Winner'] == 'South'])
sa_drawn_matches = len(df_sa[df_sa['Winner'] == 'Draw'])
sa_non_drawn_matches = sa_total_matches - sa_drawn_matches
sa_win_percentage = (sa_home_wins / sa_non_drawn_matches) * 100

print("Total Home Matches Played:", sa_total_matches)
print("Total Non Drawn Matches:", sa_non_drawn_matches)
print("Total Home Wins:", sa_home_wins)
print("Win Percentage:", sa_win_percentage, "%")
```

```python
teams = ['India', 'England', 'Australia', 'New Zealand', 'South Africa']
win_percentages = [ind_win_percentage, eng_win_percentage, aus_win_percentage, nz_win_percentage, sa_win_percentage]
average_win_percentage = np.mean(win_percentages)

plt.figure(figsize=(10,6))

bars = plt.bar(teams, win_percentages, color=['orange', 'blue', 'gold', 'black', 'green'], alpha=0.7)


plt.axhline(y=average_win_percentage, color='red', linestyle='--')

plt.title("Home Win Percentages Across Teams", fontsize=16)
plt.xlabel("Teams", fontsize=14)
plt.ylabel("Home Win Percentage", fontsize=14)

plt.grid(True, axis='y', linestyle='--', alpha=0.5)
plt.ylim(45, 75)

plt.show()
```
